```python
#production of the crops from the year 2006-07
plt.figure(figsize=(24,10))
sns.barplot(x="Crop",y="Production 2006-07", data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x="Crop",y="Area 2006-07", data=df, marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="Crop",y="Yield 2006-07",marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
# comparison between production,area,yeild for the crops from year 2006-07

plt.figure(figsize=(20,7))
sns.barplot(x="Crop",y="Production 2006-07", data=df, color='lightblue')
sns.lineplot(x="Crop",y="Area 2006-07", data=df, marker='o',label='Area 2006-07')
sns.lineplot(x="Crop",y="Yield 2006-07", data=df, marker='o',label='Yield 2006-07')
plt.xticks(rotation=90)
```
